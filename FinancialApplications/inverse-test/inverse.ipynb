```python
import csv, math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def inverse(data):
    result = []
    for i in range(len(data)):
        result.append(1/data[i])
    return result
```

```python
df_apple = pd.read_csv('../data/apple_prices.csv', names=['Price'], header=None)
apple = df_apple.Price.tolist()[:200]
```

```python
df_apple_inv = pd.read_csv('../data/prices_inv_plain.csv', names=['Price'], header=None)
apple_inv = df_apple_inv.Price.tolist()
```

```python
df_noises = pd.read_csv('../data/white_noise.csv', names=['Noise'], header=None)
noises = df_noises.Noise.tolist()[:200]
```

```python
df_noises_inv = pd.read_csv('../data/noises_inv_plain.csv', names=['Noise'], header=None)
noises_inv = df_noises_inv.Noise.tolist()
```

```python
df_small_noises = pd.read_csv('../data/white_noise_small.csv', names=['Noise'], header=None)
small_noises = df_small_noises.Noise.tolist()
```

```python
df_small_noises_inv = pd.read_csv('../data/small_noises_inv_plain.csv', names=['Noise'], header=None)
small_noises_inv = df_small_noises_inv.Noise.tolist()
```

```python
apple_inv_exact = inverse(apple)
```

```python
noises_inv_exact = inverse(noises)
```

```python
small_noises_inv_exact = inverse(small_noises)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, figsize = (20, 30))

ax1.plot(apple_inv_exact, label='Exact Inverse')
ax1.plot(apple_inv, label='Approximated Inverse')
ax1.set_title('Apple Stock Prices Inverse Test')
ax1.legend()

ax2.plot(noises_inv_exact, label='Exact Inverse')
ax2.plot(noises_inv, label='Approximated Inverse')
ax2.set_title('White Noises Inverse Test')
ax2.legend()

ax3.plot(small_noises_inv_exact, label='Exact Inverse')
ax3.plot(small_noises_inv, label='Approximated Inverse')
ax3.set_title('Small White Noises Inverse Test')
ax3.legend()

plt.show()
```
