```python
from scipy import interpolate
import matplotlib.pyplot as plt
from numpy import linspace
```

```python
# x = [-4, -3, -2, -1, 0, 1, 2, 3, 4]
# y = [0, 0, 0, 0, 1, 0, 0, 0, 0]
x = [-2, -1, 0, 1, 2]
y = [0, 0, 1, 0, 0]
```

```python
f_1 = interpolate.BarycentricInterpolator(x, y)
x_1 = linspace(-2, 2, 200)
y_1 = f_1(x_1)
```

```python
plt.figure(figsize=(20,10))
plt.plot(x, y, 'o', x_1, y_1, '-')
plt.show()
```

```python
ls = [1, 1, -1, 1, 0, 1, 1, 1, -1]
for i in range(len(ls)):
    a = 0
    for j in range(len(ls)):
        tmp = ls[i]-ls[j]
        a += f_1(tmp)
    print(a)
```
